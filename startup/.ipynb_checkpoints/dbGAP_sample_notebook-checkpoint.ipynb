```python
# John's function for selecting the best calling threshold per region
from scipy.stats import beta
from core.modeling_utilities.misc import list_like

class ChoseSettingsAndRegionsForHMF():
    def __init__(self,
                 pessimistic_effect_size_ci: float = 0.90,
                 low_noise_cutoff_percentile_per_region: float = 90,
                 min_prob_clean: float = 0.75,
                 min_effect_size: float = 1.0,
                 desired_noise_floor: float = 1 / 1500,
                 use_minimal_noise_floor_no_obs: bool = False,
                 minimal_noise_floor_with_no_obs: float = 1 / 500,
                 use_fold_change: bool = True,
                 min_features: int = 2):
        """ Given a samples x regions x settings x (n hypermethyl, n total) dimensional count object, this pretrainer
            learns the best setting per region in the fit function, and selects that per sample in the transform.
            This results in a cleaner samples x regions x (n hypermethyl, n total) we can use as input to a downstream
            outlier detector, or could be transformed into a vector of hypermethylation rates downstream. Additionally
            this pretrainer will remove any region where there is no setting that passes low noise in healthy and high
            effect size in healthy vs cancer samples.

        Args:
            train_samples: list of Sample objects to use to determine which of y == 1 are more progressed
                (eg stage 4 and larger tumor mm)
            min_stage_or: minimum stage for a sample to be included on the y==1 side (an or option)
            min_effect_size: minimum effect size of the final set of regions after chosing the "best" setting per region
            min_tumor_size_mm_or: minimum tumor size to be included on the y==1 side (an or option)
            pessimistic_effect_size_ci: the confidence interval to use in the definition of pessimistic effect size
            low_noise_cutoff_percentile_per_region: noise needs to be low, higher values are more stringent 0-100
            min_prob_clean: the minimum probability a region has a hypermethylation rate under the noise floor
            desired_noise_floor: the noise floor for the hypermethylation rate
            use_minimal_noise_floor_no_obs: If there are no positive read observations, use a smaller noise floor to
                allow in more regions with no clear evidence of noise.
            minimal_noise_floor_with_no_obs: The alternative noise floor to use in the case that
                `use_minimal_noise_floor_no_obs` is `True`
            add_pon: whether to query and add PON samples to the healthy side in the evaluation
            add_high_tf: whether to query and add known high tumor fraction samples to the cancer side in the evaluation
            use_fold_change: Use fold change as a criteria in determining optimal HMF settings per region
            use_all_samples_effect_size: use all samples in computing effect size, not just the ones that pass minimal
                tumor size or stage thresholds.
            case_class_label: The case class label, defaults to 1, which should be appropriate until we do multiclass.
            min_features: minimum number of features to pass through, will take the best
                n even if they do not pass cutoffs
            pon_class: the class name of the PON to query (only used if add_pon is True)
            high_tf_class: the class name of the high tf samples to query (only used if add_high_tf is True)
            feature_name: the feature name to use for the pon_class and/or the high_tf_class, if needed
        """
        self.desired_noise_floor = desired_noise_floor
        self.pessimistic_effect_size_ci = pessimistic_effect_size_ci
        self.min_prob_clean = min_prob_clean
        self.min_effect_size = min_effect_size
        self.low_noise_cutoff_percentile_per_region = low_noise_cutoff_percentile_per_region
        
        self.use_fold_change = use_fold_change
        self.min_features = min_features
        self.minimal_noise_floor_with_no_obs = minimal_noise_floor_with_no_obs
        self.use_minimal_noise_floor_no_obs = use_minimal_noise_floor_no_obs
        self.features_to_keep = None
        self.prob_lt_noise_floor = None
        self.prob_lt_noise_floor_all = None
        self.pessimistic_effect_size_all = None
        self.pessimistic_effect_size = None
        self.best_feature_per_region = None
        self.prob_lt_minimal_noise_floor = None
        self.positive_observations_healthy = None
        self.total_reads_healthy = None
        self.passing_noise_floor = None

    # return the input data array (x) subset to the best threshold per region and only
    # markers above the noise floor
    def select_best_features_per_region(self, x, included_sample_dim=True):
        """Does the kind of odd feature selection we have to do with numpy indexing"""
        if included_sample_dim:
            return x[:, np.arange(x.shape[1]), self.best_feature_per_region]
        else:
            return x[np.arange(x.shape[0]), self.best_feature_per_region]

    def fit(self, case_samples, control_samples, **fit_params):
        """ Fits the background expectation over
        Args:
            x (numpy array): data on which to fit
            y (numpy array): labels to define groups
        Returns:
        """
        
        case_totals = np.sum(case_samples, axis=0)

        assert control_samples.shape[0] >= 2, "Too few control samples for fit."
        control_for_clean_fit = np.sum(control_samples, axis=0)

        # Fit a beta-binomial distribution over possible underlying ratios
        
        # for each region, create a cdf of a beta random variable based on control samples. 
        # fit beta.cdf(noise_floor, number methyl frags, total frags)
        prob_lt_noise_floor_all = beta.cdf(
            self.desired_noise_floor,
            control_for_clean_fit[..., 0] + 1,  # number of positive observations
            control_for_clean_fit[..., 1] - control_for_clean_fit[..., 0] + 1  # negative = total - positive
        )
        prob_lt_noise_floor_minimal_no_noise_all = beta.cdf(
            self.minimal_noise_floor_with_no_obs,
            control_for_clean_fit[..., 0] + 1,  # number of positive observations
            control_for_clean_fit[..., 1] - control_for_clean_fit[..., 0] + 1  # negative = total - positive
        )

        # should be regions x settings in shape
        self.prob_lt_noise_floor_all = prob_lt_noise_floor_all
        
        # Endpoints of the range that contains self.pessim percent of the distribution
        # highest methyl ratio expected in controls
        _, control_high = beta.interval(
            self.pessimistic_effect_size_ci,
            control_for_clean_fit[..., 0] + 1,  # number of positive observations
            control_for_clean_fit[..., 1] - control_for_clean_fit[..., 0] + 1  # negative = total - positive
        )
        # lowest methyl ratio expected in cases
        case_low, _ = beta.interval(
            self.pessimistic_effect_size_ci,
            case_totals[..., 0] + 1,  # number of positive observations
            case_totals[..., 1] - case_totals[..., 0] + 1  # negative = total - positive
        )

        # should be regions x settings in shape
        self.pessimistic_effect_size_all = case_low / control_high

        # Goal: per region find the index of the best feature (best calling threshold) by low noise and high effect size
        _best_feature_per_region = []
        for effect_size_i, prob_lt_noise_floor_i in zip(self.pessimistic_effect_size_all, self.prob_lt_noise_floor_all):
            # get a reasonable cutoff for a low-ish noise level for this region either top n percentile, or passing
            # our low end cutoff
            if self.use_fold_change:
                idx = np.arange(len(effect_size_i))  # save the indices because we need to subset them by noise passing
                low_noise_cutoff = np.percentile(prob_lt_noise_floor_i, self.low_noise_cutoff_percentile_per_region)
                noise_passing_i = (prob_lt_noise_floor_i >= low_noise_cutoff) | \
                                  (prob_lt_noise_floor_i >= self.min_prob_clean)
                not_nan_effect_size_i = ~np.isnan(effect_size_i)
                # find the index of the setting for this feature that is:
                #  1) has healthy noise less than or equal to the noise cutoff defined above
                #  2) has the biggest effect size conditional on this lower noise level
                # get the index offset of the best effect size conditioned on noise passing
                passing_i_noise_and_not_nan = noise_passing_i & not_nan_effect_size_i
                if np.sum(passing_i_noise_and_not_nan) == 0:
                    # fall back to the max prob less than noise floor if effect size is all nan
                    logger.warning("All of the noise passing sites were NaN for effect size, "
                                   "ignoring effect size and choosing the best settings by clean probability.")
                    best_i = np.nanargmax(prob_lt_noise_floor_i)
                else:
                    argsort_effect_size_i = np.argsort(effect_size_i[passing_i_noise_and_not_nan])
                    best_i = idx[passing_i_noise_and_not_nan][argsort_effect_size_i[-1]]
            else:
                # if we only use the healthy samples for this calculation, then simply take the argmax for the highest
                #  probability that the ratio is under the desired noise floor.
                #  at the time this code was written, the argmax function returns the single smallest index of the max
                #  given ties. This is the desired behavior
                best_i = np.nanargmax(prob_lt_noise_floor_i)
            if list_like(best_i):
                best_i = best_i[0]  # take the first element in the list if multiple bests
            _best_feature_per_region.append(best_i)

        self.best_feature_per_region = np.array(_best_feature_per_region)

        self.prob_lt_noise_floor = self.select_best_features_per_region(prob_lt_noise_floor_all,
                                                                        included_sample_dim=False)

        self.positive_observations_healthy = self.select_best_features_per_region(
            control_for_clean_fit[..., 0],
            included_sample_dim=False,
        )

        self.total_reads_healthy = self.select_best_features_per_region(
            control_for_clean_fit[..., 1],
            included_sample_dim=False,
        )

        self.prob_lt_minimal_noise_floor = self.select_best_features_per_region(
            prob_lt_noise_floor_minimal_no_noise_all,
            included_sample_dim=False,
        )

        self.pessimistic_effect_size = self.select_best_features_per_region(self.pessimistic_effect_size_all,
                                                                            included_sample_dim=False)
        # find out which features we want to keep
        self.passing_noise_floor = self.prob_lt_noise_floor >= self.min_prob_clean
        if self.use_minimal_noise_floor_no_obs:
            self.passing_noise_floor = (
                    self.passing_noise_floor |
                    ((self.prob_lt_minimal_noise_floor >= self.min_prob_clean) & (
                            self.positive_observations_healthy == 0))
            )
        if self.use_fold_change:
            # best min_features by probability less than noise x effect size
            top_n_probs = np.argsort(self.prob_lt_noise_floor * self.pessimistic_effect_size)[-self.min_features:]
            self.features_to_keep = self.passing_noise_floor & (self.pessimistic_effect_size >= self.min_effect_size)
        else:
            # best min_features by probability less than noise
            top_n_probs = np.argsort(self.prob_lt_noise_floor)[-self.min_features:]
            self.features_to_keep = self.passing_noise_floor
        self.features_to_keep[top_n_probs] = True  # set the best n to true as well as ones that pass cutoff
        return self

    def transform(self, x):
        """ Calcualtes the log10 fold change of observed values to the expected values given depth
        Args:
            x (numpy array): data to transform
        Returns:
            x (numpy array): transformed data
        """
        return self.select_best_features_per_region(x, included_sample_dim=True)[:, self.features_to_keep]
```

```python
# a function to return the data matrix of region counts using the lowest calling threshold and noise floor
# to call fragments

def return_best_counts(mat, 
                       case_classes: list, 
                       control_classes:list):
    
    """
    Use John's functions to return the methyl counts/total counts ratio data for 
    samples from a set of case classes and control classes. Extracts the counts for only regions
    that have a signal above the noise floor (calculated using a beta-binomial for each region) 
    using the best read calling threshold for the comparison of cases and controls
    above a certain average effect size.
    
    mat: ndarray of samples x regions x thresholds x (methylcounts, counts)
    """
    
    assert len(case_classes) >= 1
    assert len(control_classes) >= 1

    # label each sample as one of the case classes or not
    case_sel = None
    for case_class in case_classes:
        bools = mat.get_labels() == name_to_label[case_class]
        if case_sel is None:
            case_sel = bools
        else:
            case_sel = case_sel | bools

    # label each sample as one of the control classes or not
    control_sel = None
    for control_class in control_classes:
        bools = mat.get_labels() == name_to_label[control_class]
        if control_sel is None:
            control_sel = bools
        else:
            control_sel = control_sel | bools

    case_x = mat.x[case_sel]
    control_x = mat.x[control_sel]
    
    all_labels = [label_to_name[cls] for cls in mat.get_labels()]
    case_labels = [label for label in all_labels if label in case_classes]
    control_labels = [label for label in all_labels if label in control_classes ]

    # initialize and fit data object. 
    case_control = ChoseSettingsAndRegionsForHMF()
    case_control = case_control.fit(case_samples=case_x, control_samples=control_x)
    col_meta = mat.get_column_metadata()[case_control.features_to_keep]
    
    # return cases and control counts using the best threshold feature per region
    case_x_tfm, control_x_tfm = case_control.transform(case_x), case_control.transform(control_x)
    
    # calculate ratio of methyl counts to total counts
    case_ratios = case_x_tfm[..., 0] / case_x_tfm[..., 1]
    control_ratios = control_x_tfm[..., 0] / control_x_tfm[..., 1]
    
    ratios = np.concatenate((case_ratios, control_ratios), axis=0)
    labels = case_labels + control_labels
    
    return ratios, labels, col_meta
```

```python
# Instead select regions at a threshold of 3 CpGs with > NUM total counts in all samples
label_to_name = {
    5:'CRC_tissue',
    4:'AA_CAP',
    3:'AA_CFP',
    2:'Normal_Colon_Tissue', 
    1:'PBL_CAP',
    0:'PBL_CFP',
}


def counts_by_threshold(mat, NREADS):
    """Return matrix of methyl/total count ratios using CpG threshold = 3,  for regions where there are at 
    least NREADS total reads.
    """
    # plot how sparse is the RRBS data for the 434 v1 panel regions
    data_thr3 = mat.x[:,:,0,1]
    region_sums_across_samples = np.sum(data_thr3, axis=0)
    plt.figure()
    plt.hist(region_sums_across_samples, bins=100)
    plt.title(f"Sum of total fragment counts per region {data_thr3.shape[1]}")
    plt.ylabel("Number of regions")
    plt.xlabel("Total fragment counts")
    #plt.xlim((0,300))

    regions_to_keep = []
    for col in range(data_thr3.shape[1]):
        if not any(data_thr3[:,col] < NREADS):
            regions_to_keep.append(col)

    
    mat3 = mat.x[:,:,0,:] # counts at thresh cpg = 3
    mat4 = mat3[:,regions_to_keep,:] # filter to regions with >= NREADs in all samples
    ratios_sel_reg = mat4[:,:,0] / mat4[:,:,1]
    col_meta_sel_reg = mat.get_column_metadata()[regions_to_keep]

    # 367 regions have enough reads in the RRBS data
    print(ratios_sel_reg.shape)
    
    return ratios_sel_reg, col_meta_sel_reg, regions_to_keep
```

```python
# A functionto perform clustering
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def run_clusters(mat, classes, figname):
    
    class_set = sorted(list(set(classes)))

    pca = PCA(n_components=4)
    pca = pca.fit(mat)
    ratios_r = pca.transform(mat)
    #print(ratios_r)

    # how similar are normal tissue samples in the new space?
    #samples = [label=="Normal_Colon_Tissue" for label in labels]
    #print(ratios_r[samples,:])

    # Plot PCA and LDA results
    colors = ['purple', 'turquoise', 'darkorange', 'pink', 'cyan', 'blue']
    colors = colors[0:len(class_set)]
    lw = 2

    plt.figure(figsize=(10,10))

    ## PCA Plot
    plt.subplot(221)
    for color, cls in zip(colors, class_set):
        samples = [name == cls for name in classes]
        plt.scatter(ratios_r[samples, 0], ratios_r[samples, 1], color=color, alpha=.8, lw=lw,
                    label=cls)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.xlabel(f"PC1 ({round(pca.explained_variance_ratio_[0]*100,1)}%)")
    plt.ylabel(f"PC2 ({round(pca.explained_variance_ratio_[1]*100,1)}%)")
    plt.title('PCA of RRBS Data')

    plt.subplot(222)
    for color, cls in zip(colors, class_set):
        samples = [name == cls for name in classes]
        plt.scatter(ratios_r[samples, 2], ratios_r[samples, 3], color=color, alpha=.8, lw=lw,
                    label=cls)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.xlabel(f"PC3 ({round(pca.explained_variance_ratio_[2]*100,1)}%)")
    plt.ylabel(f"PC4 ({round(pca.explained_variance_ratio_[3]*100,1)}%)")
    plt.title('PCA of RRBS Data')


    ## LDA Plot
    if len(class_set) < 3: 
        print("Too few classes for doing LDA")
    else:
        lda = LinearDiscriminantAnalysis(n_components=len(class_set)-1)
        lda = lda.fit(mat, classes)
        ratios_r2 = lda.transform(mat)
        #print(ratios_r2)
        
        plt.subplot(223)
        for color, cls in zip(colors, class_set):
            samples = [name == cls for name in classes]
            plt.scatter(ratios_r2[samples, 0], ratios_r2[samples, 1], color=color, alpha=.8, lw=lw,
                        label=cls)
        plt.legend(loc='best', shadow=False, scatterpoints=1)
        plt.xlabel(f"LDA1 ({round(lda.explained_variance_ratio_[0]*100,1)}%)")
        plt.ylabel(f"LDA2 ({round(lda.explained_variance_ratio_[1]*100,1)}%)")
        plt.title('LDA of RRBS Data')

        if len(class_set) >= 3:
            plt.subplot(224)
            for color, cls in zip(colors, class_set):
                samples = [name == cls for name in classes]
                plt.scatter(ratios_r2[samples, 1], ratios_r2[samples, 2], color=color, alpha=.8, lw=lw,label=cls)
            plt.legend(loc='best', shadow=False, scatterpoints=1)
            plt.xlabel(f"LDA2 ({round(lda.explained_variance_ratio_[1]*100,1)}%)")
            plt.ylabel(f"LDA3 ({round(lda.explained_variance_ratio_[2]*100,1)}%)")
            plt.title('LDA of RRBS Data')

    plt.savefig(f"./{figname}")
    
    return lda
```

## Cluster the dbGAP tissue samples using differences between blood and tissue. Use this feature space to look for clustering in the cfDNA DC1 samples. 

```python
# extract paired sample data

meta = mat2.get_metadata()
names = np.array([label_to_name[cls] for cls in mat2.get_labels()])
patient_ids = np.array([str(dict['public_metadata']['submitted_subject_id']) for dict in meta])

#print(crosstab(patient_ids, names))

pbl_samples = np.array([ ("PBL" in x) for x in names])
blood_data = mat2.x[pbl_samples,...]
blood_patient_ids = patient_ids[pbl_samples]

# a subset of 20 patients have blood samples
#print(len(blood_patient_ids))

# extract patient samples with matching blood samples
filt = [(id in blood_patient_ids) for id in patient_ids]
paired_data = mat2.x[filt,:,0,:]
paired_ids = patient_ids[filt]
paired_names = names[filt]
#print(paired_data.shape) #three dimensions

#filter out regions without at least NUM fragments in all samples
NUM = 10
regions_to_keep = []
for col in range(paired_data.shape[1]):
    if not any(paired_data[:,col,1] < NUM):
        regions_to_keep.append(col)
paired_data = paired_data[:,regions_to_keep,:]
paired_data_ratios = paired_data[...,0] / paired_data[...,1]

#print(paired_data_ratios.shape) #two dimensions

# calculate tissue-blood ratio diffs
filt = [("PBL" in x) for x in paired_names]
blood_ratios = paired_data_ratios[ filt,:]
blood_ids = paired_ids[filt]
blood_names = paired_names[filt]

filt = [("PBL" not in x) for x in paired_names]
tissue_ratios = paired_data_ratios[filt,:]
tissue_ids = paired_ids[filt]
tissue_names = paired_names[filt]
tissue_combos = [f"{x}:{y}" for x,y in zip(tissue_ids, tissue_names)]

# for each tissue sample, calc diff with blood
patient_diffs = []
diff_names = []
diff_patients = []
for combo in set(tissue_combos):
    patient = combo.split(":")[0]
    tissue_class = combo.split(":")[1]
    blood_val = blood_ratios[blood_ids==patient][0]  # a 1D array
    
    filt = [x==combo for x in tissue_combos]
    tissue_val = tissue_ratios[filt]  # a 2D array

    # if more than one tissue type replicate in the patient take the avg vals
    if tissue_val.shape[0] > 1:
        tissue_val = np.mean(tissue_val, axis=0)  # a 1D array
    
    if len(tissue_val.shape) > 1:
        tissue_val = tissue_val[0,:] # a 1D array
    
    diff = tissue_val - blood_val  # a 1D array
    patient_diffs.append(diff.tolist())
    diff_names.append(tissue_class)
    diff_patients.append(patient)

#print(patient_diffs[0])
df_diffs = pd.DataFrame.from_records(patient_diffs)
df_diffs.index = [f"{x}:{y}" for x,y in zip(diff_patients, diff_names)]
print(df_diffs.shape)
df_diffs.head()
```

```python
# cluster the diff data by tissue type

lda = LinearDiscriminantAnalysis(n_components=len(set(diff_names))-1)
lda = lda.fit(df_diffs, diff_names)
ratios = lda.transform(df_diffs)

plt.figure(figsize=(20,10))
plt.subplot(121)
for color, cls in zip(colors, set(diff_names)):
    samples = [name == cls for name in diff_names]
    plt.scatter(ratios[samples, 0], ratios[samples, 1], color=color, alpha=.8, lw=lw,
                label=cls)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.xlabel(f"LDA1 {round(lda.explained_variance_ratio_[0]*100,1)}")
plt.ylabel(f"LDA2 {round(lda.explained_variance_ratio_[1]*100,1)}")

plt.subplot(122)
for color, cls in zip(colors, set(diff_names)):
    samples = [name == cls for name in diff_names]
    plt.scatter(ratios[samples, 1], ratios[samples, 2], color=color, alpha=.8, lw=lw,label=cls)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.xlabel(f"LDA2 {round(lda.explained_variance_ratio_[1]*100,1)}")
plt.ylabel(f"LDA3 {round(lda.explained_variance_ratio_[2]*100,1)}")
```

```python
# how similar are the predictors in pbl vs in cfDNA 
# plot the rank of predictors in PBL_CAP/PBL_CFP comparison vs rank in 
"""
dc_classes = [('colorectal_benign_targetmethyl1_discovery1p5', 5),
                 ('colorectal_cancer_targetmethyl1_discovery1p5', 4),
                 ('colorectal_healthy_targetmethyl1_discovery1p5', 3),
                 ('colorectal_healthy_targetmethyl1_discovery1p5_panelofneg', 2), 
                 #('all_unclassifiable_targetmethyl1_discovery1p5', 1),
                 ('colorectal_precancer_targetmethyl1_discovery1p5', 0)]
            
dc_label_to_name = {5:"Benign",
                   4:"Cancer",
                   3:"Healthy",
                   2:"PoN",
                   0:"Pre_Cancer"} 
"""
from scipy.stats import mannwhitneyu, rankdata, pearsonr
from sklearn.linear_model import LinearRegression

# calculate mat_rrbs and mat_dc for the same set of regions first . . .
def run_rank_corr(dc_control_classes, dc_case_classes, figname):
    """Are similar sets of genes DMR between our classes and between the PBL classes? 
    Run MWU tests between cases and controls for methylation ratios in each region of the DC data set. 
    Run MWU tests between controls (PBL_CFP) and cases (PBL_CAP) in the RRBS data set.
    Plot and measure correlation between ranks in the two comparisons"""
    
    assert mat434_ratios.shape[1] == dc_ratios.shape[1], "PBL and DC data sets dont have the same number of regions"
    
    # filter dc data to classes of interest
    samples = [ cls in (dc_control_classes + dc_case_classes) for cls in dc_names]
    dc_ratios2 = dc_ratios[samples]
    dc_classes2 = np.array(dc_classes)[samples]
    
    # reclassify cases and controls in the DC data set
    dc_classes2 = [ "case" if cls in dc_case_classes else "control" for cls in dc_classes2 ]
    
    # for each region in the rrbs data set calc MWU
    rrbs_mwu_statistics = []
    rrbs_mwu_pvals = []
    dc_mwu_statistics = []
    dc_mwu_pvals = [] 
    
    for col in range(mat434_ratios.shape[1]):
        
        # compare PBL_CFP and PBL_CAP
        pbl_cap = mat434_ratios[[(cls == "PBL_CAP") for cls in mat434_classes]][:,col]
        #print(pbl_cap.shape)
        pbl_cfp = mat434_ratios[[(cls == "PBL_CFP") for cls in mat434_classes]][:,col]
        #print(pbl_cfp.shape)
        if not any([x > 0 for x in (list(pbl_cap) + list(pbl_cfp))]):
            #print(f"region with no PBL methylation at all: {col}")
            U_pbl, p_val_pbl = 0, 1.0
        else:
            U_pbl, p_val_pbl = mannwhitneyu(pbl_cfp, pbl_cap, alternative='two-sided')
        
        
        # compare DC control and case
        dc_control = dc_ratios2[[(cls == "control") for cls in dc_classes2]][:,col]
        #print(dc_control.shape)
        dc_case = dc_ratios2[[(cls == "case") for cls in dc_classes2]][:,col]
        #print(dc_case.shape)
        U_dc, p_val_dc = mannwhitneyu(dc_control, dc_case, alternative='two-sided')
        
        # use only significantly DMR regions
        p_thresh = 0.1
        if (p_val_pbl < p_thresh):
            dc_mwu_statistics.append(U_dc)
            dc_mwu_pvals.append(p_val_dc)
            rrbs_mwu_statistics.append(U_pbl)
            rrbs_mwu_pvals.append(p_val_pbl)
    
    #print(dc_mwu_pvals)
    #print(rrbs_mwu_pvals)
    #plt.figure()
    #plt.hist(rrbs_mwu_pvals)
    #plt.figure()
    #plt.hist(dc_mwu_pvals)
    
    #plot rank correlation by p_val
    plt.figure(figsize=(20,5))
    plt.subplot(121)
    x = rankdata(rrbs_mwu_pvals)
    y = rankdata(dc_mwu_pvals)
    x_arr = x.reshape(-1,1)
    y_arr = y.reshape(-1,1)
    lm = LinearRegression()  
    lm.fit(x_arr, y_arr)  
    Y_pred = lm.predict(x_arr)  
    plt.plot(x_arr, Y_pred, color='red', ls="-")
    plt.scatter(x, y)
    corr, corr_p = pearsonr(x, y)
    plt.xlabel("PBL DMR significance rank")
    plt.ylabel("DC DMR significance rank")
    plt.title(f"p-value {dc_control_classes} vs. {dc_case_classes}")
    plt.text(x=max(x)-4, y=max(y), s=f"corr = {np.round(corr,2)}")
    plt.text(x=max(x)-4, y=max(y)-2, s=f"corr_p = {np.round(corr_p,2)}")
    
    # plot rank correlation by U statistic
    plt.subplot(122)
    x = rankdata(rrbs_mwu_statistics)
    y = rankdata(dc_mwu_statistics)
    x_arr = x.reshape(-1,1)
    y_arr = y.reshape(-1,1)
    lm = LinearRegression()  
    lm.fit(x_arr, y_arr)  
    Y_pred = lm.predict(x_arr)  
    plt.plot(x_arr, Y_pred, color='red', ls="-")
    plt.scatter(x, y)
    corr, corr_p = pearsonr(x, y)
    plt.xlabel("PBL DMR significance rank")
    plt.ylabel("DC DMR significance rank")
    plt.title(f"U-statistic {dc_control_classes} vs. {dc_case_classes}")
    plt.text(x=max(x)-4, y=max(y), s=f"corr = {np.round(corr,2)}")
    
    plt.savefig(f"./{figname}"")
    
    
```

```python
"""
dc_label_to_name = {5:"Benign",
                   4:"Cancer",
                   3:"Healthy",
                   2:"PoN",
                   0:"Pre_Cancer"} 
"""

dc_control_classes = ["Benign"]
dc_case_classes = ["Pre_Cancer"]
run_rank_corr(dc_control_classes=dc_control_classes, dc_case_classes=dc_case_classes)

dc_control_classes = ["Benign"]
dc_case_classes = ["Pre_Cancer", "Cancer"]
run_rank_corr(dc_control_classes=dc_control_classes, dc_case_classes=dc_case_classes)

dc_control_classes = ["Benign"]
dc_case_classes = ["Cancer"]
run_rank_corr(dc_control_classes=dc_control_classes, dc_case_classes=dc_case_classes)

dc_control_classes = ["Benign", "Pre_Cancer"]
dc_case_classes = ["Cancer"]
run_rank_corr(dc_control_classes=dc_control_classes, dc_case_classes=dc_case_classes)

dc_control_classes = ["Healthy", "PoN"]
dc_case_classes = ["Cancer"]
run_rank_corr(dc_control_classes=dc_control_classes, dc_case_classes=dc_case_classes)

```
