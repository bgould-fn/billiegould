```python
import pandas as pd 
import numpy as np 


from lims_api import Client
#lims_url = 'http://research-lims.freenome.net'
lims_url = 'http://intercluster'
c = Client(lims_url, )

import matplotlib.pyplot as plt
# import seaborn as sns
from scipy.stats import chi2_contingency, mannwhitneyu
from featureio import FeatureIO
```

```python
### starting point = query copied from lims web ui:

tubes = [114087, # got this from the study7 manifest
114274,
115845,
119381,
121445,
122035,
123643,
125141,
243998,
243999,
244000,
244001,
244002,
244003,
244004,
244005,
244006,
244007,
244008,
244009,
244010,
244011,
244013,
244014,
244015,
244016]


query = {
    "and": [
        {
            "sample_datasets": [
                [
                    "tube_id",
                    "in",
                    tubes
                ]
            ]
        }
    ]
}

results = c.search(query , load_models = ['samples', 'patients', 'diagnoses'], 
                           query_model = 'sample_datasets')
results = results['sample_datasets']
```

```python
ds_ids = [x['id'] for x in results]
sample_ids = [x['sample_id'] for x in results]
patient_ids = [x['sample']['patient']['id'] for x in results]
```

```python
df = pd.DataFrame([ds_ids, sample_ids, patient_ids]).T
df.columns = ['ds_id', 'sample_id', 'pt_id']
```
